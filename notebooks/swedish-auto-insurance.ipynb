### Imports, Loading Data, and Visualizing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/swedish_insurance.csv")
df.plot(kind="scatter", x="X", y="Y", color="red")

plt.title("Number of Claims Vs. Money Paid")
plt.xlabel("Number of Claims")
plt.ylabel("Money Paid")
```

```python
# Calculating mean, variance, and covariance

mean_x = np.mean(df['X'])
mean_y = np.mean(df['Y'])

var_x = np.var(df['X'])
var_y = np.var(df['Y'])

print("X stats: mean=%.3f variance=%.3f" % (mean_x, var_x))
print("Y stats: mean=%.3f variance=%.3f" % (mean_y, var_y))

covar_matrix = np.cov(df['X'], df['Y'])
covar_xy = covar_matrix[0][1]
print(covar_xy)
```

```python
# Estimating coefficients
from sklearn.metrics import r2_score

b1 = covar_xy / var_x
b0 = mean_y - b1 * mean_x

x = df['X'].values.copy()

y_hat = b1 * x + b0

x_range = pd.Series(range(0, 124))
y_predicted = b1 * x_range + b0

df.plot(kind="scatter", x='X', y='Y')
plt.plot(x_range, y_predicted)

y_hat = b1 * df['X'] + b0

r2 = r2_score(y_hat, df['Y'])
print(r2)

```
